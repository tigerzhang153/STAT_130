To create a 95% bootstrapped confidence interval using the bootstrapped sample means (without using their standard deviation), you rank the bootstrapped means in ascending order. Then, you select the 2.5th percentile and the 97.5th percentile of the bootstrapped means. These percentiles form the lower and upper bounds of the 95% confidence interval, respectively. This method directly uses the distribution of the bootstrapped means to construct the confidence interval, rather than relying on the standard error.

Work with a ChatBot to create code to produce a 95% bootstrap confidence interval for a population mean based on a sample that you have and comment the code to demonstrate how the code can be changed to produce a 95% bootstrap confidence interval for different population parameter (other than the population mean, such as the population median).<br>

````python
import numpy as np

# Sample data (replace this with your actual sample)
sample = np.array([10, 20, 30, 40, 50])

# Number of bootstrap samples
n_bootstraps = 1000

# Set up an empty list to store the bootstrapped sample means
bootstrapped_means = []

# Create bootstrap samples and compute their means
for _ in range(n_bootstraps):
    # Resample from the sample with replacement
    boot_sample = np.random.choice(sample, size=len(sample), replace=True)
    
    # Compute the mean of the bootstrap sample
    boot_mean = np.mean(boot_sample)
    
    # Store the bootstrap mean
    bootstrapped_means.append(boot_mean)

# Compute the 2.5th and 97.5th percentiles to get the 95% confidence interval
lower_bound = np.percentile(bootstrapped_means, 2.5)
upper_bound = np.percentile(bootstrapped_means, 97.5)

# Print the confidence interval for the mean
print(f"95% Bootstrap Confidence Interval for the Mean: ({lower_bound}, {upper_bound})")
````

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Read the data from a CSV file
data = pd.read_csv("alitech_vaccine_data.csv")

# Calculate the difference in health scores
data['HealthScoreDifference'] = data['FinalHealthScore'] - data['InitialHealthScore']

# Set the random seed for reproducibility
np.random.seed(42)

# Bootstrapping function
def bootstrap_means(data, n_iterations=10000):
    boot_means = []
    for _ in range(n_iterations):
        sample = data.sample(frac=1, replace=True)
        mean_difference = sample['HealthScoreDifference'].mean()
        boot_means.append(mean_difference)
    return np.array(boot_means)

# Run bootstrapping
boot_means = bootstrap_means(data)

# Plot the distribution of bootstrap means
plt.hist(boot_means, bins=30, edgecolor='black')
plt.title('Bootstrap Distribution of Mean Differences')
plt.xlabel('Mean Difference in Health Scores')
plt.ylabel('Frequency')

# Save the plot as a PNG file
plt.savefig("bootstrap_distribution.png", format='png')

# Optionally, display the plot (remove this if you just want to save the image)
plt.show()

# Calculate the observed mean difference
observed_mean_diff = data['HealthScoreDifference'].mean()

# Calculate the 95% confidence interval
conf_interval = np.percentile(boot_means, [2.5, 97.5])

print(f"Observed Mean Difference: {observed_mean_diff}")
print(f"95% Confidence Interval: {conf_interval}")
````
